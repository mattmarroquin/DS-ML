```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

dataMale = pandas.read_csv('insurance male.csv')
dataFemale = pandas.read_csv('insurance female.csv')
```

```python
dataMale.describe()
```

```python
dataFemale.describe()
```

```python
male_x = DataFrame(dataMale, columns=['age'])
male_y = DataFrame(dataMale, columns=['charges'])

female_x = DataFrame(dataFemale, columns=['age'])
female_y = DataFrame(dataFemale, columns=['charges'])
```

```python
plt.figure(figsize=(10,10))
plt.scatter(male_x, male_y, alpha=0.4)
plt.title('male age vs healthcare costs')
plt.xlabel('male age')
plt.ylabel('charges')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(female_x, female_y, alpha=0.4)
plt.title('female age vs healthcare costs')
plt.xlabel('female age')
plt.ylabel('charges')
plt.show()
```

```python
regressionMale = LinearRegression()
regressionMale.fit(male_x, male_y)
```

```python
regressionFemale = LinearRegression()
regressionFemale.fit(female_x, female_y)
```

```python
regressionMale.coef_
```

```python
regressionFemale.coef_
```

```python
regressionMale.intercept_
```

```python
regressionFemale.intercept_
```

```python
plt.figure(figsize=(10,10))
plt.scatter(male_x, male_y, alpha=0.4)

# Adding the regression line here:
plt.plot(male_x, regressionMale.predict(male_x), color='orange', linewidth=3)
plt.title('male age vs healthcare costs')
plt.xlabel('male age')
plt.ylabel('charges')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(female_x, female_y, alpha=0.4)

# Adding the regression line here:
plt.plot(female_x, regressionFemale.predict(female_x), color='orange', linewidth=3)
plt.title('female age vs healthcare costs')
plt.xlabel('female age')
plt.ylabel('charges')
plt.show()
```

The R-Squared scores show that when examining a wide range of charges, female healthcare costs are more explained by age than that of men. 

```python
regressionMale.score(male_x, male_y)
```

```python
regressionFemale.score(female_x, female_y)
```
