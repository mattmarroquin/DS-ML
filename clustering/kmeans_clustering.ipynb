```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import copy
import numpy as np



#dataset
file = pd.read_csv("hwk08.csv")
df = pd.DataFrame(file)

#min and max values of columns
maxA = df['A'].max()
maxB = df['B'].max()
maxC = df['C'].max()
minA = df['A'].min()
minB = df['B'].min()
minC = df['C'].min()


def plotkmeans(col1, col2, mx1, mx2, mn1, mn2, numclusters):
    #get the largest value and smallest value of the two columns used
    maximum = max(mx1, mx2)
    minimum = min(mn1, mn2)
    
    #centroids
    centroids = {
        i+1: [np.random.randint(minimum, maximum), np.random.randint(minimum, maximum)]
        for i in range(numclusters)
    }

    #centroid code
    kmeans = KMeans(n_clusters=numclusters)
    kmeans.fit(df[[col1, col2]])

    labels = kmeans.predict(df[[col1, col2]])
    centroids = kmeans.cluster_centers_

    #plotting code
    fig = plt.figure(figsize=(8, 8))
    colmap = {1: 'r', 2: 'g', 3: 'b', 4: 'c', 5: 'm', 6: 'y', 7: 'k', 8: 'w'}

    colors = list(map(lambda x: colmap[x+1], labels))

    plt.scatter(df[col1], df[col2], color=colors, alpha=0.5, edgecolor='k')
    for idx, centroid in enumerate(centroids):
        plt.scatter(*centroid, color=colmap[idx+1])
    plt.xlim(mn1-1, mx1+1)
    plt.ylim(mn1-1, mx2+1)
    plt.show()
```

```python
plotkmeans('A', 'B', maxA, maxB, minA, minB, 3)
```

```python
plotkmeans('A', 'B', maxA, maxB, minA, minB, 5)
```

```python
plotkmeans('A', 'C', maxA, maxC, minA, minC, 8)
```
